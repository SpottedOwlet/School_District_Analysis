NaN means not a number and can not be equal to zero.

#### Options for Handling Missing Data:

##### Option1: Do nothing
This would not impact sum or averaging as NaN values are ignored during these operations. However, multiplication or division with a row that has NaN, will give asnwers as NaN.
##### Option2: Drop the Row
Dropping rows can affect the story you are trying to tell with the data. Before removing rows with NaN, you should ask yourself two key questions:
How much data would be removed if NaNs are dropped?
How would this impact the analysis?

to drop the NaNs
missing_grade_df.dropna()

##### Option3: Fill in the Row
In Pandas, we use the fillna() method to fill in a row. If we want to fill all empty rows with zero in our DataFrame, we pass the "0" in parentheses like this: 
df.fillna(0)

```python
# Drop the NaNs
missing_grades_df.dropna()
```

```python
# Fill in empty rows with values
missing_grades_df.fillna(85)
```
